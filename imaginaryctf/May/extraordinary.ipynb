````python
def xor(a,b) -> bytes:
    return bytes([
        x^y for x,y in zip(bytes(a),bytes(b))
    ])
    
def bor(a,b) -> bytes:
    return bytes([
        x|y for x,y in zip(bytes(a),bytes(b))
    ])
````
